Perhaps the column "8" is intended to change the sign of the column "6".

```python
dtf3 = dtf.copy()
dtf3["6"] = abs(dtf3["6"])
print(dtf3["6"])
```

```python
corr_matrix = dtf3.corr(method="pearson")
plt.figure(figsize = (30,30))
sns.heatmap(corr_matrix, vmin=-1., vmax=1., annot=True, fmt='.2f', cmap="YlGnBu", cbar=True, linewidths= 0.5)
plt.title("pearson correlation")
```

And now colomn "6" have very strong correlation with "target".

```python
x = dtf3['6']
y = dtf3['target']
plt.figure(figsize=(6, 6))
plt.plot(x, y)
```

It looks like the formula of "target" is x^2 + z. And if we look at the table we see that x is "6" and z is all another.


Let's check if it is true.

```python
dtf4 = dtf.assign(Z= dtf["target"] - dtf["6"]*dtf["6"])
dtf4.head(20)
```

```python
corr_matrix = dtf4.corr(method="pearson")
plt.figure(figsize = (30,30))
sns.heatmap(corr_matrix, vmin=-1., vmax=1., annot=True, fmt='.2f', cmap="YlGnBu", cbar=True, linewidths= 0.5)
plt.title("pearson correlation")
```

We see that colomn "7" is similar to colomn "z". So my hypothesis is true.
And that mean: target = "6" * "6" + "7".

Verification

```python
dtf["Prediction"] = dtf["6"]*dtf["6"]+dtf["7"]

count = 0
for i in range(len(dtf)):
    if dtf["target"].iloc[i] >= dtf["Prediction"].iloc[i] - 0.0001:
        if dtf["target"].iloc[i] <= dtf["Prediction"].iloc[i] + 0.0001:
            count += 1
        
print(count)
```
